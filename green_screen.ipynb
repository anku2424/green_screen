```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('images/girl.jpg')

image_copy = np.copy(image)

image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
print (image_copy.shape)
plt.imshow(image_copy)
```

```python
green_low = np.array([0,150,0])
green_high = np.array([130,255,130])
```

```python
# Now we will mask the car from the image using cv2's inRange function

mask = cv2.inRange(image_copy, green_low, green_high)
plt.imshow(mask, cmap='gray')
```

```python
# Now extracting the car from the green image

masked_image = np.copy(image_copy)
masked_image[mask != 0] = [0,0,0]

plt.imshow(masked_image)
```

```python
space_background_image = cv2.imread('images/nature1.jpg')
space_background_copy = cv2.cvtColor(space_background_image, cv2.COLOR_BGR2RGB)
plt.imshow(space_background_copy)
```

```python
nature_masked_image = np.copy(space_background_copy)
nature_masked_image = nature_masked_image[nature_masked_image.shape[0]-332:, nature_masked_image.shape[1]-590:]
print (nature_masked_image.shape)
nature_masked_image[mask == 0] = np.array([0,0,0])
plt.imshow(nature_masked_image)
```

```python
final_image = nature_masked_image + masked_image

plt.imshow(final_image)
plt.imsave("images/final_img", final_image)
```
